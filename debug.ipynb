```python
import time
import re
import glob
import requests
import json

import pandas as pd

from urllib.request import urlopen
from urllib import request

from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import Select,WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By

options = Options()
options.add_argument('--headless')    # ヘッドレスモードに
driver = webdriver.Chrome()
wait = WebDriverWait(driver, 10)


URL = "https://tsukubadtm.bandcamp.com/"
driver.get(URL)
time.sleep(1)
```

```python
element = driver.find_elements(By.CLASS_NAME, "leftMiddleColumns")
# driver.execute_script("arguments[0].click();", element)
```

```python
arr = []

url_list = []
eplist = element[0].find_elements(By.TAG_NAME, "li")[:]
for e in eplist:
    url     = e.find_elements(By.TAG_NAME, "a")[0].get_attribute('href')
    ep_title = e.find_elements(By.TAG_NAME, "a")[0].text
    url_list.append([url, ep_title])
```

```python
arr = []
for (url, ep_title) in url_list:
    time.sleep(1)

    # EPのページにアクセス
    driver.get(url)
    xpath = "//script[@type='application/ld+json']"
    script = driver.find_elements(By.XPATH, xpath)

    c = script[0].get_attribute('innerHTML').replace("\n", "")
    d = json.loads(c)

    album_id = str(d["albumRelease"][0]["additionalProperty"][0]["value"])
    image_url = d["albumRelease"][0]["image"][0]

    df = pd.DataFrame()
    for i, v in enumerate(d["track"]["itemListElement"]):
        track_id   = str(v["item"]["additionalProperty"][0]["value"])
        music_name = v["item"]["name"]
        
        try:
            artist_name = v["item"]["byArtist"]["name"]
        except:
            artist_name = None

        df.loc[i, "album_id"]    = album_id
        df.loc[i, "album_name"]  = ep_title
        df.loc[i, "image_url"]   = image_url
        df.loc[i, "track_id"]    = track_id
        df.loc[i, "music_name"]  = music_name
        df.loc[i, "artist_name"] = artist_name
    arr.append(df)

    print(ep_title, url)

pd.concat(arr).reset_index(drop=1).to_csv("csv/musiclist.csv")
```

```python
pd.concat(arr).reset_index(drop=1).to_csv("csv/musiclist.csv", encoding="utf-8 sig", index=False)
```
